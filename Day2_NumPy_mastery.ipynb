```python
'''
1 数组创建 
任务：用3种不同方式创建数组：
1) 从列表 [1,2,3,4,5]；
2) 创建一个3x3的全1矩阵；
3) 创建一个形状为(2,4)，值从0到7的数组。 
目标：掌握 np.array(), np.ones(), np.arange().reshape()。 
'''
import numpy as np 
#创建列表
list1 = np.array([1,2,3,4,5])
print(list1)
#创建一个3x3的全1矩阵
list2 = np.ones((3,3))
print(list2)
#创建0-7数组
list3 = np.arange(0,8,1)
print(list3)
#改变形状
rlist3 = np.reshape(list3,(2,4))
print(rlist3)
```

```python
'''
2 索引与切片 
任务：创建一个4x4数组（0-15），然后：
1) 取出第2行；2) 取出第1列；
3) 取出中心2x2的子矩阵（即行1-2，列1-2）；
4) 使用布尔索引取出所有大于10的元素。 
目标：理解二维数组的 array[row, col] 索引方式。
'''
import numpy as np 
#创建4*4数组
arr = np.arange(0,16,1)
print(arr)
rarr = np.reshape(arr,(4,4))
print(rarr)
#取出第2行
rarr1 = rarr[1]
print(rarr1)
#取出第1列
rarr2 = rarr[:,0]#不能省略冒号
print(rarr2)
#取出中心2*2的子矩阵
rarr3 = rarr[0:2,0:2]
print(rarr3)
#取出大于10的元素
rarr4 = rarr[rarr>10]
print(rarr4)
```

```python
'''
3 形状操作 
任务：将一个1维数组 np.arange(12) 变形为：
1) 3行4列的矩阵；
2) 变回1维；
3) 使用 np.newaxis 将其从(12,)变为(12,1)的列向量。 
目标：理解 .reshape(), .flatten(), np.newaxis。
'''
import numpy as np 
arr1 = np.arange(12).reshape(3,4)
print(arr1)
arr2 = arr1.flatten()#数组方法，不能写np.flatten(arr1)
print(arr2)
arr3 = arr2[:,np.newaxis]#冒号不能省略
print(arr3)
```

```python
'''
4 数学运算 
任务：创建两个2x3的随机矩阵A和B，计算：
1) 对应元素相加、相乘；
2) 矩阵乘法（注意维度）；
3) 对A求所有元素的和、每列的平均值、每行的最大值。 
目标：区分 * 和 @，掌握聚合函数。
'''
import numpy as np 
A = np.random.rand(2,3)
B = np.random.rand(2,3)
print(A)
print(B)
#对应元素相加
C = A + B 
print(C)
#对应元素相乘 
D = A * B 
print(D)
#矩阵乘法
E = np.dot(A,B.T)
print(E)
#对A求所有元素的和、每列的平均值、每行的最大值。
Asum = np.sum(A)
colAve = np.mean(A,axis = 0)
rowMax = np.max(A,axis = 1)
print(f"A所有元素和：{Asum}")
print(f"A每列平均值：{colAve}")
print(f"A每行最大值：{rowMax}")
```

```python
'''
· 背景：线性回归试图找到一条直线 y = w * x + b 来拟合数据。
· 你的任务：
  1. 生成模拟数据：用 np.linspace 创建100个在0-10之间的x值，用 y = 2 * x + 1 + 噪声 生成y值（噪声用 np.random.randn）。
  2. 计算最优参数：使用正规方程（Normal Equation）直接计算最优的 w 和 b。公式为：
     
     weights = (X^T * X)^(-1) * X^T * y 
     
     其中，X 是一个100x2的矩阵，第一列为x值，第二列全为1（对应截距b）。
  3. 可视化结果（可选但推荐）：用 matplotlib 绘制原始数据散点图和你拟合出的直线。
'''
import numpy as np
import matplotlib.pyplot as plt

#1.生成模拟数据
np.random.seed(42)
x = np.linspace(0,10,100)
y_true = 2*x + 1
y_noise = np.random.randn(100)*2
y = y_true + y_noise

#2.计算最优参数
#（1）构造X矩阵,将x增加一列全1
xMatrix = np.column_stack((x, np.ones_like(x)))
#(2) 计算
w =np.linalg.inv (xMatrix.T @ xMatrix) @ xMatrix.T @ y

#3.可视化结果
slope,intercept = w[0],w[1] #斜率、截距
print(f"拟合结果：斜率 = {slope:.3f}，截距 = {intercept:.3f}")
print(f"真实参数：斜率 = 2.000，截距 = 1.000")
#创建图形和坐标轴
plt.figure(figsize = (10,6)) #设置图形大小：宽10英寸，高6英寸
#绘制原始数据散点图
plt.scatter(x,y,
           alpha = 0.6,   #透明度
           s = 20,        #点的大小
           c = 'steelblue',   #点的颜色
           edgecolors = 'black',   #点边缘颜色
           linewidths = 0.5,   #边缘线宽
           label = '原始数据（含噪声）')
#绘制真实关系直线（红色虚线）
plt.plot(x,y_true,
        'r--',   #红色虚线
        linewidth = 2,
        alpha = 0.8,
        label = '真实关系：y = 2x + 1')
#绘制拟合的直线（绿色实线）
y_pred = slope*x + intercept
plt.plot(x,y_pred,
           'g-',
           linewidth = 3,
           alpha = 0.8,
           label = f'拟合直线：y = {slope:.2f}x + {intercept:.2f}')
#添加图例和标签
plt.legend(loc = 'upper left',fontsize = 12) #图例位置
plt.xlabel('x值',fontsize = 14)
plt.ylabel('y值',fontsize = 14)
plt.title('NumPy实现线性回归与Matplotlib可视化',fontsize = 16,pad = 20)
#添加网格线
plt.grid(True,alpha = 0.3)
#自动调整布局并显示图形
plt.tight_layout()
plt.show()
```
